```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from itertools import permutations, combinations 
from sklearn import datasets
```

```python
class my_LDA():
    def __init__(self):
        self.classes = np.array([0])
        self.class_counts = []
        self.mean_global = np.array([0])
        self.corvariance_global = np.array([0])
        self.probability_vector = []
        self.means = []
        self.groups = []
        self.covariance_sets = []
    
    def fit(self, X, y):
        self.classes = np.unique(y)
        
        #  
        self.mean_global = np.mean(X, axis=0)
         
        for cls in self.classes:
            x = X[y==cls]
            self.probability_vector.append( len(x)/ len(X))
            self.class_counts.append(len(x))
             
            #  (x)  
            self.means.append(np.mean(x, axis=0) )
            
            #     (  )
            x_d = x-self.mean_global
            self.groups.append(x_d)
            
            #   
            v = x_d.T.dot(x_d)  
            self.covariance_sets.append(v)
         
        # lda    , 
        self.corvariance_global = np.zeros(self.covariance_sets[0].shape)
        for idx, cv in enumerate(self.covariance_sets):
            self.corvariance_global += cv / len(y) 
       
        #  
        self.corvariance_global = np.linalg.inv(self.corvariance_global)
        
        
    def predict(self, X):
        
        res = []
        #        
        for idx, cls in enumerate(self.classes): 
            pread = self.means[idx].dot(self.corvariance_global).dot(X.T) -\
            self.means[idx].dot(self.corvariance_global).dot(self.means[idx])*0.5 +\
            np.log(self.probability_vector[idx])
        
            res.append(pread)
        
        res = np.vstack(res).T
        
        return res 
    
iris = datasets.load_iris()
X = iris['data'][:, (2, 3)]
y = (iris['target']==2).astype(int)
lda = my_LDA()
lda.fit(X, y)

XX, YY = np.meshgrid(np.linspace(0, 7, 100), np.linspace(0, 3, 100))
x_new = np.c_[XX.ravel(), YY.ravel()]
pred = lda.predict(x_new)
ZZ = np.argmax(pred, axis=1).reshape(XX.shape)

cmap = mpl.colors.ListedColormap(sns.color_palette(["r", "g"]).as_hex())
plt.contourf(XX, YY, ZZ, alpha=0.5, cmap=cmap)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap) 
```

```python
from itertools import permutations, combinations 
class my_Multiple_LDA():
    def __init__(self): 
        #   OvO lda 
        self.combinations = []
        self.classifiers = []
        self.classes = []
    
    def fit(self, X, y): 
        self.classes = list(np.unique(y))
        
        self.combinations = combinations(self.classes, 2)
         
        for cls in self.combinations:  
            x_ = X[np.isin(y, cls)]
            y_ = y[np.isin(y, cls)] 
            
            lda = my_LDA()
            lda.fit(x_, y_) 
            
            self.classifiers.append(lda)
            
    def predict(self, X):
        
        candidates = dict([(k, 0) for k in self.classes])
        t=[]
        for lda in self.classifiers:
            t.append(lda.predict(X))
    
 
        summary = []
        for idx, a in enumerate(t):
            
            winner_ = np.argmax(a, axis=1) 
            winner_ = np.array(self.classifiers[idx].classes[winner_])
 
            summary.append(winner_)
    
        summary = np.vstack(summary).T
  
        res = []
        for cls in range(0, 3):
            r = (summary==cls).astype(int).sum(axis=1)
            res.append(r)

        res = np.vstack(res).T
        
        res = np.argmax(res, axis=1) 
        return res
         
    
iris = datasets.load_iris()
X = iris['data'][:, (2, 3)]
y = iris['target']
lda = my_Multiple_LDA()
lda.fit(X, y)


XX, YY = np.meshgrid(np.linspace(0, 7, 100), np.linspace(0, 3, 100))
x_new = np.c_[XX.ravel(), YY.ravel()]
ZZ = lda.predict(x_new).reshape(XX.shape) 

cmap = mpl.colors.ListedColormap(sns.color_palette(["r", "g", 'b']).as_hex())
plt.contourf(XX, YY, ZZ, alpha=0.5, cmap=cmap)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap) 
```
